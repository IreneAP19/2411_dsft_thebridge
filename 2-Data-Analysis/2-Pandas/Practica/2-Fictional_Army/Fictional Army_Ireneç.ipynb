### Step 4. Fill missing values with the mean of their regiment.

```python
#Hay que sustituir los valores Nan por la media de, en este caso readiness, cada regimiento
#tendremos que hacer 3 sub DF, uno por regimineto, y hacer la media de la columna para poder sustituir el valor
```

```python
army
```

```python
m1=army[army["regiment"]== "Nighthawks"]["readiness"].mean()

m2=army[army["regiment"]== "Dragoons"]["readiness"].mean()

m3=army[army["regiment"]== "Scouts"]["readiness"].mean()
m3,m2,m1
```

```python
l1=army[army["regiment"]== "Nighthawks"].fillna(m1)
l2=army[army["regiment"]== "Dragoons"].fillna(m2)
l3=army[army["regiment"]== "Scouts"].fillna(m3)
pd.concat([l1,l2,l3],ignore_index=True)
```
