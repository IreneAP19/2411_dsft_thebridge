24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
#mujeres OR hombres EACH clase
counted = pd.pivot_table(titanic,index="gender",columns="pclass",values="survived",aggfunc="sum").T#variable de survived
counted
#la tasa de supervivencia se saca con la media
#count contaria todos los pasajero pero quiero solo los que sobrevivieron, por eso uso sum
```

34. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.<>

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

![kdeplot](./img/kdeplotsur.png)

```python
sns.kdeplot(data=titanic[titanic["survived"]==0][["age"]], x="age",color="red",label="No Survived",fill=True);#cogeme las edades de los muertos del titanic, en este caso age no es necesario, ya que damos la mascara y luego la columna que queremos (age)
sns.kdeplot(data=titanic[titanic["survived"]==1], x="age",color="green",label="Survived",fill=True);
plt.legend()#la legend esta en la propia visualizacion(label)
```

```python
filtro = titanic[titanic["survived"]==0][["age"]]#pongo doble corchete para que no me cre una serie, sino que coja las edades para cada no superviviente
sns.kdeplot(data=filtro, x="age",color="red",label="No Survived",fill=True);
#nos dice la probabiidad de la edad mas probable que obtengamos si cogemos una persona aleatoria
```
