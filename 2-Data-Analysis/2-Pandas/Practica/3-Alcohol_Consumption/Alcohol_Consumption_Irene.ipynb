### Step 4. Which continent drinks more beer on average?

```python
#drinks[drinks["continent"]== null].fillna("NA")(inplace=True)
```

```python
#agrupo los continentes y las cervezas y hago la media del consumo de cerveza por continente
x=drinks.groupby("continent")["beer_servings"].mean()
```

```python
x.idxmax()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
vino=drinks.groupby("continent")["wine_servings"].describe()
vino
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
medias=drinks.groupby("continent")[["wine_servings","beer_servings","spirit_servings","total_litres_of_pure_alcohol"]].mean()
medias
#Usamos dos [] para seleccionar varias columnas RECORDAR
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
mediana=drinks.groupby("continent")[["wine_servings","beer_servings","spirit_servings","total_litres_of_pure_alcohol"]].median()
mediana
```

### Step 8. Print the mean, min and max values for spirit consumption per continent.

```python
estudio=drinks.groupby("continent")["spirit_servings"].aggregate(['mean','min', 'max'])
estudio
```
